```python
# numpy is used for matrix calculation
import numpy as np
# tensorflow is used to build machine learning models
import tensorflow as tf
# matplotlib is used to plot graphs
import matplotlib.pyplot as plt
```

```python
# Start creating dataset
np.random.seed(42)
X = np.random.standard_normal((200,1)) # rows, columns
```

```python
X
```

```python
# Define function
y = 2 * X + 3
```

```python
# Plot function
plt.plot(X,y)
plt.grid(True)
```

```python
# Define the Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=[1])
])
model.summary()
```

```python
# Plot Model
tf.keras.utils.plot_model(model, show_shapes=True)
```

```python
# Build Model
model.compile(loss=tf.keras.losses.mean_squared_error, optimizer=tf.keras.optimizers.Adam())
# Train Model
history = model.fit(x=X, y=y, epochs=1000)
```

```python
# Get predicted y
y_pred = model.predict(X)
# Compare real y and predicted y
np.c_[y, y_pred]
```

```python
# Find the loss
tf.keras.losses.mean_squared_error(y[:,0], y_pred[:,0])
```

```python
# Plot Loss History
loss_hist = history.history['loss']
plt.plot(range(1000), loss_hist)
plt.grid(True)
```
